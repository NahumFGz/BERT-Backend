```
models      = ['SpanBERT', 'BETO', 'multilingual']
datasets    = ['Total_15_19', 'Total_20_24', 'Total_25_29']
balanceos   = ['normal', 'over', 'under']
test_sz     = ['0.1','0.2', '0.3']
lrs         = ['5e-5', '3e-5', '2e-5']
num_epochs  = ['5','6','7','8','9','10'] 
flg_preproc = ['original', 'processed']

i = 1
for model in models:
  for dataset in datasets:
    for balanceo in balanceos:
      for size in test_sz:
        for lr in lrs:
          for num_epoch in num_epochs:
            for flg in flg_preproc:
              # Track experiment
              print ('Experiment N°' + str(i))

              with open('/content/BERT-Backend-Flask/zAutomate_TrainMultiBERT.py', 'r') as file:
                # read a list of lines into data
                data = file.readlines()
              
                # now change the 2nd line, note that you have to add a newline
                data[1] =  'model = ' + chr(39)+ model + chr(39) + '\n'
                data[2] =  'dataset_name = ' + chr(39)+ dataset + chr(39) + '\n'
                data[3] =  'balanceo = ' + chr(39)+ balanceo + chr(39) + '\n'
                data[4] =  'test_sz = ' + size  + '\n'
                data[5] =  'learning_rate = ' + lr  + '\n'
                data[6] =  'num_epochs = ' +  num_epoch + '\n'
                data[7] =  'flg_preproc = ' + chr(39)+ flg + chr(39) + '\n'
              
                # and write everything back
                with open('/content/BERT-Backend-Flask/zAutomate_TrainMultiBERT.py', 'w') as file:
                    file.writelines( data )
              
              #Execute Experiment
              !python /content/BERT-Backend-Flask/zAutomate_TrainMultiBERT.py

              i += 1
```
